```python
from sklearn.datasets import make_s_curve

X, color = make_s_curve(n_samples=1500, noise=0.05, random_state=42)

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(X[:, 0],  X[:, 2], c=color, cmap=plt.cm.Spectral)
plt.show()
```

```python
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt

# Generate the Swiss Roll dataset
X, color = make_swiss_roll(n_samples=1500, noise=0.8, random_state=42)

# Plot the 2D projection
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 2], c=color, cmap=plt.cm.viridis)
plt.colorbar()
plt.title('2D Projection of Swiss Roll')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4*np.pi, 1000)
r = t + np.random.normal(0, 0.1, 1000)
x = r * np.cos(t)
y = r * np.sin(t)

plt.figure(figsize=(8, 8))
plt.plot(x, y, 'b.')
plt.title('Gaussian Spiral')
plt.axis('equal')
plt.show()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X, y = make_circles(n_samples=1000, factor=0.5, noise=0.05)

plt.figure(figsize=(8, 8))
plt.scatter(X[y==0, 0], X[y==0, 1], c='purple', label='Inner circle')
plt.scatter(X[y==1, 0], X[y==1, 1], c='maroon', label='Outer circle')
plt.title('Concentric Circles')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate the moons dataset
X, y = make_moons(n_samples=1000, noise=0.1, random_state=42)

# Visualize the moons
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Moons Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Class')
plt.show()
```
